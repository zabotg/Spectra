```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

## Path correlations

```python
path_corr = "/home/zabot/Documents/Codes/Mestrado/Spectra/src/result-correlation.txt"
```

```python
correlations = pd.read_csv(path_corr, sep=';', header=None)
correlations.columns = ["Dist-1", "Dist-2", "FEM-1", "FEM-2", "Correlation"]
correlations
```

```python
euclidean = correlations.loc[(correlations['Dist-1'] == 'euclidean') & (correlations['Dist-2'] == 'euclidean')]
labels = euclidean['FEM-1'] + '-' + euclidean['FEM-2']

plt.figure(figsize=(18,8))
plt.plot(labels, euclidean['Correlation'], label='Euclidean-Euclidean')
plt.xticks(rotation=90)
plt.legend()
plt.grid()
plt.show()
```

```python
distance_1 = pd.unique(correlations['Dist-1'])
distance_2 = pd.unique(correlations['Dist-2'])

plt.figure(figsize=(18,8))

for dist1 in distance_1:
    for dist2 in distance_2:
        result = correlations.loc[(correlations['Dist-1'] == dist1) & (correlations['Dist-2'] == dist2)]
        plt.plot(range(result['Correlation']), result['Correlation'])
    plt.show()
```
