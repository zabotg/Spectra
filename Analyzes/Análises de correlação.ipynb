```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
```

## Path correlations

```python
path_corr = "/home/zabot/Documents/Codes/Mestrado/Spectra/src/result-correlation.txt"
```

```python
correlations = pd.read_csv(path_corr, sep=';', header=None)
correlations.columns = ["Dist-1", "Dist-2", "FEM-1", "FEM-2", "Correlation"]

# create a new column 
correlations['Combined'] = correlations['FEM-1'] + '-' + correlations['FEM-2']
print(correlations['Dist-1'].unique())
```

### Correlation graph

```python
plt.figure(figsize=(20,11))

correlations.sort_values(by=['Correlation'], inplace=True)
sns.boxplot(x='Combined', y='Correlation', data=correlations)

plt.axvspan(xmin=-1, xmax=91, ymin=0, ymax=0.36, alpha=0.1, color='green') #Low
plt.axvspan(xmin=-1, xmax=91, ymin=0.36, ymax=0.7, alpha=0.1, color='blue') #Medium
plt.axvspan(xmin=-1, xmax=91, ymin=0.7, ymax=1.2, alpha=0.1, color='red') #High

plt.xticks(rotation=90)
plt.xlabel('Combination of Pairs of FEMs')
plt.title('Correlation variation with the exchange of function pairs of distances')
plt.grid()
plt.tight_layout()
plt.savefig('correlationVariation.pdf', transparent=True)
plt.show()
```
